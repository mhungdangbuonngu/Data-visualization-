```python
import numpy as np
import pandas as pd
```

```python
with open("C:/Users/daoth/Downloads/Homework_sampledata1.txt",'r') as f:
    lines = f.readlines() 
lines=lines[1:]
for i in range(len(lines)):
    lines[i]=lines[i].split()[1:]
for i in range(len(lines)):
    for j in range(len(lines[i])):
        lines[i][j]=float(lines[i][j])

rain_1st_day=lines[0]
first_day_data=[[None]*3 for i in range(len(rain_1st_day))]
for i in range(0,len(rain_1st_day)):
    if i < 9:
        first_day_data[i][0]='2008-0{month}-01'.format(month=i+1)
    else:
        first_day_data[i][0]='2008-{month}-01'.format(month=i+1)
    first_day_data[i][1]='00:00:00'
    first_day_data[i][2]=rain_1st_day[i]
first_day_df=pd.DataFrame(first_day_data)
first_day_data
```

```python
all_day=[[None]*3 for i in range(len(lines)*len(lines[0]))]
for i in range(len(lines[0])):
    for j in range(len(lines)):
        if i < 9 and j<9:
            all_day[i*len(lines)+j][0]='2008-0{month}-0{day}'.format(month=i+1,day=j+1)
        elif i< 9 and j >=9:
            all_day[i*len(lines)+j][0]='2008-0{month}-{day}'.format(month=i+1,day=j+1)
        elif i>=9 and j<9:
            all_day[i*len(lines)+j][0]='2008-{month}-0{day}'.format(month=i+1,day=j+1)
        elif i>=9 and j>=9:
            all_day[i*len(lines)+j][0]='2008-0{month}-{day}'.format(month=i+1,day=j+1)
        all_day[i*len(lines)+j][1] = '00:00:00'
        if lines[j][i]==-99:
            all_day[i*len(lines)+j][2]=None
        else:
            all_day[i*len(lines)+j][2]=lines[j][i]
all_day_df=pd.DataFrame(all_day)
all_day_df
```

```python
# Drop the wrong date
for i in range(len(all_day_df)):
    if all_day_df[0][i] == '2008-02-30' or all_day_df[0][i] == '2008-02-31' or all_day_df[0][i] == '2008-04-31' or all_day_df[0][i] == '2008-06-31' or all_day_df[0][i] == '2008-09-31' or all_day_df[0][i] == '2008-11-31':
        all_day_df = all_day_df.drop(i)
all_day_df
```

```python
all_day_list=all_day_df[[0,2]].values.tolist()
all_day_list
```

```python
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(len(all_day_list)):
    x.append(all_day_list[i][0])
    y.append(all_day_list[i][1])
plt.figure(figsize=(20,5))
plt.plot(x,y,color='blue')
plt.xticks([0, 31, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335], ["01\n Jan 2008", "01\n Feb", "01\n Mar", "01\n Apr", "01\n May", "01\n Jun", "01\n Jul", "01\n Aug", "01\n Sep", "01\n Oct", "01\n Nov", "01\n Dec"])
plt.grid(True)
plt.ylim(5,35)
plt.xlim(0,365)
plt.xlabel("time")
plt.ylabel('temperature')
plt.title('temperature in 2008')
plt.legend(['temperature'])
plt.savefig('Figure1_1.png')
plt.show()
```

```python
import math
all_days_list_copy = all_day_list.copy()
# Calculate the average temperature of each month
for i in range(len(all_days_list_copy)):
    if math.isnan(all_days_list_copy[i][1]):
        all_day_list[i][1] = 0
avg_temp = [None for i in range(12)]
num_day_month = 0
num_day_prev = 0
for i in range(12):
    avg_month = 0
    num_day_prev = num_day_prev + num_day_month 
    if i == 1: 
        # February in 2008 has 29 days
        num_day_month = 29
    elif i == 3 or i == 5 or i == 8 or i == 10:
        num_day_month = 30
    else:
        num_day_month = 31
    num_day_cal = num_day_month
    for j in range(num_day_month):
        # Calculate the average temperature of each month
        if all_day_list[num_day_prev+j][1] == 0:
            # If the temperature of a day is missing, we will not calculate it
            num_day_cal = num_day_cal - 1
        avg_month += all_day_list[num_day_prev+j][1]
    avg_temp[i] = avg_month/num_day_cal
# Plot the graph of average temperature of each month
plt.figure(figsize=(20,5))
plt.grid(True)
plt.ylim(10, 28)
plt.xlim(-0.5, 11.5)
plt.plot(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov","Dec"], avg_temp, color='blue')
plt.plot(avg_temp, 'o', color='blue')
plt.xlabel('Month')
plt.ylabel('Temperature(oC)')
plt.title('Average temperature of each month in 2008')
plt.legend(['Temperature'])
plt.savefig('Figure1_2.png')
plt.show()
```
